## Traducir al Ingl√©s los comentarios

```python
def translate_opinion(opinion):
    return GoogleTranslator(source='auto', target='en').translate(opinion)
```

```python
tqdm.pandas()
input_r['Reviews'] = input_r['Opini√≥n'].progress_apply(translate_opinion)
```
