# An√°lisis de Sentimientos

### Modelo VADER

```python
sia = SentimentIntensityAnalyzer()
```

```python
aprobacion1 = {}

for i, row in tqdm(input_v.iterrows(), total=len(input_v)):
    text = row['Reviews']
    mid = row['√çndice']
    aprobacion1[mid] = sia.polarity_scores(text)
```

```python
SenAn = pd.DataFrame(aprobacion1).T
SenAn = SenAn.reset_index().rename(columns={'index':'√çndice','neg':'Vader negativo','neu':'Vader neutral','pos':'Vader positivo','compound':'Valor Vader'})
SenAn
```

```python
input_v = input_v.merge(SenAn, how='left')
input_v
```

```python
V_vader = sns.barplot(data=input_v, x='Valoraci√≥n',y='Valor Vader')
V_vader.set_title('Restaurantes Per√∫: An√°lisis de Sentimientos en las opiniones de Google Maps', fontsize='xx-large')
plt.show()
```

```python
fig, VC_vader = plt.subplots(1, 3, figsize=(14,5))
sns.barplot(data=input_v, x='Valoraci√≥n', y='Vader positivo', ax=VC_vader[0])
sns.barplot(data=input_v, x='Valoraci√≥n', y='Vader neutral', ax=VC_vader[1])
sns.barplot(data=input_v, x='Valoraci√≥n', y='Vader negativo', ax=VC_vader[2])
VC_vader[0].set_title('Vader Positivo')
VC_vader[1].set_title('Vader Neutral')
VC_vader[2].set_title('Vader Negativo')
fig.suptitle('Restaurantes Per√∫: Comparaci√≥n por clasificaci√≥n de sentimientos', fontsize='xx-large')
plt.tight_layout()
plt.show()
```

### Modelo RoBERTa

```python
Roberta = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(Roberta)
model = AutoModelForSequenceClassification.from_pretrained(Roberta)
```

```python
def roberta_polarity_scores(text):
    encoded_text = tokenizer(text, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'Roberta negativo' : scores[0],
        'Roberta neutro' : scores[1],
        'Roberta positivo' : scores[2]
    }
    return scores_dict
```

```python
aprobacion2 = {}

for i, row in tqdm(input_v.iterrows(), total=len(input_v)):
    try:
        text = row['Reviews']
        mid = row['√çndice']
        aprobacion2[mid] = roberta_polarity_scores(text)
    except RuntimeError:
        print(f'Broke at id {mid}')
```

```python
SenARob = pd.DataFrame(aprobacion2).T
SenARob = SenARob.reset_index().rename(columns={'index':'√çndice','Roberta neutro':'Roberta neutral'})
SenARob
```

```python
input_v = input_v.merge(SenARob, how='left')
input_v
```

```python
polarity_weights = torch.tensor([-1, 0, 1])
probs = torch.tensor(input_v[["Roberta negativo", "Roberta neutral", "Roberta positivo"]].values)
polarity = polarity_weights * probs
polarity = polarity.sum(dim=-1)
polarity_scaled = nn.Tanh()(polarity)
input_v["Valor Roberta"] = polarity_scaled.numpy()
input_v["Roberta sentiment"] = pd.cut(input_v["Roberta polarity"],
    bins=[-1.0, -0.25, 0.25, 1.0],labels=["Negative_R", "Neutral_R", "Positive_R"],)
input_v.head()
```

```python
V_vader = sns.barplot(data=input_v, x='Valoraci√≥n',y='Roberta polarity')
V_vader.set_title('Restaurantes Per√∫: An√°lisis de Sentimientos en las opiniones de Google Maps (RoBERTa)', fontsize='xx-large')
plt.show()
```

```python
fig, VC_Roberta = plt.subplots(1, 3, figsize=(14,5))
sns.barplot(data=input_v, x='Valoraci√≥n', y='Roberta positivo', ax=VC_Roberta[0])
sns.barplot(data=input_v, x='Valoraci√≥n', y='Roberta neutral', ax=VC_Roberta[1])
sns.barplot(data=input_v, x='Valoraci√≥n', y='Roberta negativo', ax=VC_Roberta[2])
VC_Roberta[0].set_title('RoBERTa Positivo')
VC_Roberta[1].set_title('RoBERTa Neutral')
VC_Roberta[2].set_title('RoBERTa Negativo')
fig.suptitle('Restaurantes Per√∫: Comparaci√≥n por clasificaci√≥n de sentimientos (RoBERTa)', fontsize='xx-large')
plt.tight_layout()
plt.show()
```

### Conclusiones y Recomendaciones:
1. Al comparar los resultados del an√°lisis con el modelo VADER y el modelo RoBERTa, podemos ver que el segundo caso diferencia las valoraciones negativas y positivas de una manera m√°s √°cida; es decir, interpreta mejor el mensaje. Esto se ve con mayor claridad en el segundo y tercer gr√°fico. Por un lado, cuando la valoraci√≥n de los usuarios en el puntaje es 1, los valores negativos de RoBERTa una frecuencia m√°s alta en altas probabilidades que los valores negativos de VADER, as√≠ como que los valores positivos de RoBERTa tiene una probabilidad mucho m√°s cercana al cero que en el caso de VADER. Por otro lado, cuando la valoraci√≥n es 5, los valores positivos del modelo RoBERTa tienen una probabilidad muy cercana al 1, mientras que en VADER el an√°lisis positivo est√° m√°s distribuido alrededor del 0.5 de probabilidad
2. En el caso de RoBERTa se observa que existe cierta relaci√≥n inversamente proporcional entre la asignaci√≥n de probabilidades de que un mismo texto es positivo o negativo, lo cual implica que las predicciones del modelo tienen coherencia en la l√≥gica. Sin embargo, cuando se observa la relaci√≥n entre los VADER positvo y negativo, se nota menor claridad en la relaci√≥n de probabilidades para ambos.
3. Adicional al procedimiento desarrollado en este caso, se recomienda profundizar en la limpieza del texto, como borrar emojis o textos no completos. Por un lado, el borrar los emojis del texto puede ayudar a disminuir la distracci√≥n en la clasificaci√≥n de los modelos para que sean de mayor calidad. Por otro lado, al realizar el scraping no siempre se extrae todo el contenido; por ello, es recomendable analizar las predicciones en dichos casos para concluir si es preferible mantener o eliminar dichos textos.
